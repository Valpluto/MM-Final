```python
models = 'models'
output = 'output'
model_name  = ["MultinomialNB", "SGDClassifier", "RandomForest"]
alors = ["kraken", "tesseract"]
```

```python
from sklearn.datasets import load_files
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from joblib import dump, load
from tqdm.notebook import tqdm
import os
from sklearn import metrics
import pandas as pd
from scipy import stats
```

```python
krakens = []
tessseracts = []
ktest = load_files("{}/{}/{}".format(output, 'kraken' ,'test'))
ttest = load_files("{}/{}/{}".format(output, 'tesseract' ,'test'))

for al in alors:
    for model in model_name:
        if 'kraken' in al:
            krakens.append(load(os.path.join(models,"{}_{}.joblib".format(al,model))))
        else:
            tessseracts.append(load(os.path.join(models,"{}_{}.joblib".format(al,model))))
        
```

Kraken output 

```python
print(krakens)
```

```python
kraken_metrics = []

for al in krakens:
    predicts = al.predict(ktest.data)
    kraken_metrics.append(metrics.classification_report(ttest.target, predicts, output_dict = True))
    print(metrics.classification_report(ttest.target, predicts))
```

```python
tesseract_metrics = []

for al in tessseracts:
    predicts = al.predict(ttest.data)
    tesseract_metrics.append(metrics.classification_report(ttest.target, predicts, output_dict = True))
    print(metrics.classification_report(ttest.target, predicts))
```

```python
# Only going to anlyizse the MultinomalNB since this has the highest accuracy for both alorithms 
.73 - .49
```

```python
kraken_NB = pd.DataFrame(kraken_metrics[0])
teseract_NB = pd.DataFrame(tesseract_metrics[0])
```
